```python
print('rjan')
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset('tips')
#sns.scatterplot(tips['total_bill'],tips['tip'])
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',size='size',style='smoker')
```

```python
titanic=pd.read_csv("train (1).csv")

sns.barplot(data=titanic,x='Pclass',y='Fare')
```

```python
#sns.displot(data=titanic,['Survived'==0]['Age'])
sns.distplot(titanic['Survived'] == 0)
```

```python
sns.displot(data=titanic, x='Age', hue='Survived',kde=True,fill=True)
```

```python
# To plot the age of those who did NOT survive
sns.displot(data=titanic[titanic['Survived'] == 0], x='Age', kde=True,fill=True)
sns.displot(data=titanic[titanic['Survived'] == 1], x='Age', kde=True,fill=True)
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kind='kde', fill=True)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
#(titanic.groupby('Pclass').mean()['Survived'].plot(kind='Bar'))
titanic.groupby('Pclass')['Survived'].mean().plot(kind='bar')
```

```python
print(titanic.groupby('Sex')['Survived'].mean())
sns.clustermap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

```python
iris=sns.load_dataset('iris')

sns.pairplot(iris,hue='species')
```
